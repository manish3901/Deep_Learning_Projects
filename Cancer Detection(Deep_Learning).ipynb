### Train_test_split & Scaling data

```python
X = cancer_df.drop('benign_0__mal_1',axis=1).values
y = cancer_df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
#DO NOT FIT TEST DATA(ONLY TRAIN DATA).SCALE ONLY FEATURES NOT LABELS.
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
#30 FEATURES (add almost same nuerons in Dense layers)
```

### CREATING A MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import Adam
```

```python
#create instance of model
seq_model = Sequential()

#Add dense layers
seq_model.add(Dense(units=30,activation='relu'))
seq_model.add(Dense(units=25,activation='relu'))
seq_model.add(Dense(units=15,activation='relu'))

#output layer(output neuron count)
#for binary classification problem.
seq_model.add(Dense(units=1,activation='sigmoid'))
#sigmoid activation fn has output between o & 1.

#compile
seq_model.compile(loss='binary_crossentropy',optimizer = 'adam')

```

### TRAINING THE MODEL

```python
seq_model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600)
```

```python
#lets add losses into a dataframe
losses = pd.DataFrame(seq_model.history.history)
```

```python
losses
```

```python
losses.plot()
```

```python
#THERE'S A LOT OF OVERFITTING AS YOU CAN SEE FROM THE PLOT.LETS USE EARLY STOPPING CALLBACK TO PREVENT THIS OVERFITTING.
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#lets create a varible to store callback values
early_stops = EarlyStopping(monitor='val_loss',patience=25,mode='min',verbose=1)
```

```python
#Re-create the model for new training.Its important(just copy paste & run)
seq_model = Sequential()

#Add dense layers
seq_model.add(Dense(units=30,activation='relu'))
seq_model.add(Dense(units=25,activation='relu'))
seq_model.add(Dense(units=15,activation='relu'))

#output layer(output neuron count)
#for binary classification problem.
seq_model.add(Dense(units=1,activation='sigmoid'))
#sigmoid activation fn has output between o & 1.

#compile
seq_model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

```python
#refit the model with callback option
seq_model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[early_stops])
```

```python
#epochs stopped at 49.
```

```python
#evaluate again to see changes
model_losses = pd.DataFrame(seq_model.history.history)
model_losses.plot()
```

```python
# ITS MUCH BETTER , BUT WE CAN IMPROVE FURTHER WITH DROPOUT LAYER TO RANDOMLY TURNING OFF NEURONS.
from tensorflow.keras.layers import Dropout
```

```python
#Re-create the model for new training.Its important(just copy paste & run)
seq_model = Sequential()

#Add dense layers
seq_model.add(Dense(units=30,activation='relu'))  #rate = randomly turn off neurons in percentage of total neurons.
seq_model.add(Dropout(rate=0.5))
seq_model.add(Dense(units=25,activation='relu'))
seq_model.add(Dropout(rate=0.5))
seq_model.add(Dense(units=15,activation='relu'))
seq_model.add(Dropout(rate=0.5))

#output layer(output neuron count)
#for binary classification problem.
seq_model.add(Dense(units=1,activation='sigmoid'))
#sigmoid activation fn has output between o & 1.

#compile
seq_model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

```python
#NOTE: WE ARE USING DROPOUT ALONG WITH CALL BACK FOR REDUCING OVERFITTING
seq_model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[early_stops])
```

```python
#epochs stopped at 102
```

```python
#evaluate again to see changes
model_losses_dropout = pd.DataFrame(seq_model.history.history)
model_losses_dropout.plot()
```

```python
#both values merged close to 100 & dropping out.Its much better model
model_losses_dropout
```
